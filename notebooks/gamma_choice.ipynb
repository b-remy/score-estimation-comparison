````python
%matplotlib nbagg
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
````

Technique 2 (Other noise levels). Choose {σi}L,i=1 as a geometric progression with common ratio γ, such that 
```
Φ(√(2D)(γ − 1) + 3γ) − Φ( √(2D)(γ − 1) − 3γ) ≈ 0.5
```

From `Improved Techniques for Training Score-Based Generative Models`.

````python
def gamma_choice_function(gamma, D=2*320**2):
    return norm.cdf(np.sqrt(2*D)*(gamma-1) + 3*gamma) - norm.cdf(np.sqrt(2*D)*(gamma-1) - 3*gamma) - 0.5
````

````python
gamma = np.logspace(-4, 0, 1000)
````

````python
plt.figure()
plt.plot(gamma, gamma_choice_function(1-gamma))
plt.axhline(0, color='red')
plt.xscale('log')
plt.xlabel('1 - γ')
plt.ylabel('Target function')
plt.savefig('gamma_choice.png')
````
